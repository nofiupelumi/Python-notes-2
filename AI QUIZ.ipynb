```python
import tkinter as tk
```

```python
from tkinter import scrolledtext, messagebox
```

```python
# Setup the main window
root = tk.Tk()
root.title("AI Quiz Generator")
```

```python
# Text box for input
text_area = scrolledtext.ScrolledText(root, wrap=tk.WORD, width=40, height=10)
text_area.pack(pady=20, padx=10)
```

```python
# Button to generate questions
def generate_questions():
    input_text = text_area.get("1.0", "end-1c")
    if not input_text.strip():
        messagebox.showerror("Error", "Please enter some text to generate questions.")
        return
    questions = ask_gpt4(input_text)
    output_area.config(state=tk.NORMAL)
    output_area.delete("1.0", tk.END)
    output_area.insert(tk.END, "\n".join(questions))
    output_area.config(state=tk.DISABLED)

generate_button = tk.Button(root, text="Generate Questions", command=generate_questions)
generate_button.pack(pady=10)
```

```python
# Area to display generated questions
output_area = scrolledtext.ScrolledText(root, wrap=tk.WORD, width=40, height=10, state=tk.DISABLED)
output_area.pack(padx=10)
```

```python
Step 2: Integrating GPT-4
Function to Interact with OpenAI GPT-4:
import openai

def ask_gpt4(text):
    try:
        response = openai.Completion.create(
            engine="text-davinci-002", # Use the latest GPT-4 engine name
            prompt=f"Generate five multiple-choice questions based on the following text: {text}",
            max_tokens=1000
        )
        return [choice['text'].strip() for choice in response['choices']]
    except Exception as e:
        messagebox.showerror("API Error", str(e))
        return 
```
