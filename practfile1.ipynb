```python
def fectch_data(csv_url):
    response = requests.get(csv_url)
    if response.status_code == 200:
        return StringIO(response.text)
    else:
        raise (f'NO files status code: {response.status_code}')
        
csv_data=fectch_data(url)
csv_data
```

```python
def read_csv(csv_url):
    reader=csv.DictReader(csv_url)
    return list(reader)
organisation_data=read_csv(csv_data)
organisation_data
```

```python
#define a variable to set max no of empoyee to 0  and industry to  emppty
def get_num_emp(data):
    max_employee = 0
    max_industry = ""


    for org in data:
        employees= int(org["Number of employees"])
        if  employees > max_employee:
            max_employee=employees
            max_industry=org["Industry"]
            
    return max_industry, max_employee
            
max_industry, max_employee=get_num_emp(organisation_data)
print('The industry with highest number is :',max_industry)
print('The number of employee is :',max_employee)
```

```python
def  avg_num_emp_spec_year(data,start_year,end_year):
    filtered_organ=[org for org in data  if  int(org['Founded'])>=start_year and  int(org['Founded'])<=end_year ]
    
    #cal total employee for the spe year
    total_employee=sum(int(org['Number of employees']) for org in filtered_organ)
    org_count=len(filtered_organ)
    #avg sum of employee
    if org_count>0:
        avg_num_employee= total_employee/org_count
        return avg_num_employee
    else:
        return 0
start_year=1990
end_year=2010
    
avg_num_employee=avg_num_emp_spec_year(organisation_data,start_year,end_year)
print(f'The average number of employee  between {start_year} and {end_year}:{avg_num_employee}')
```

```python
from datetime import datetime
def recent_org_data(data):
    current_year=datetime.now().year
    
    recent_organization=[org for org in data if int(org['Founded'])>=2015  and int(org['Founded'])<=current_year]
    return recent_organization
recent_organization_d = recent_org_data(organisation_data)
print(f'The organisaton from 2015 till now are below')
for org in recent_organization_d:
    
    print(f"The name of the organisation is{org['Name']},{org['Founded']}")
```

##GET USER WITH HIGHEST POST

```python
def get_user_with_highest_posts(user_data):
    # Initialize variables to keep track of the maximum number of posts and the corresponding user
    max_posts = 0
    max_posts_user =None

    # Iterate through the list of users
    for user in user_data:
        # Assuming the user information includes a key 'posts' containing the number of posts
        posts = user.get('user-Id',  0)

        # Update max_posts and max_posts_user if the current user has more posts
        if posts > max_posts:
            max_posts = posts
            max_posts_user = user

    # Return the user with the highest number of posts
    return max_posts_user
getmode=get_user_with_highest_posts(json_file)
print(getmode)
```
