```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#generate random data
np.random.seed(42)

#create 100 random values between 0 and 2
X=2*np.random.rand(100,1)

#create y values with noise
y=4+3*X+np.random.randn(100,1)

#print(X)
model=LinearRegression()

#fit model to the data (note 'x' is independent & 'y' is dependent) make prediction using train model
model.fit(X,y)  

X_new=np.array([[0],[2]])
y_pred=model.predict(X_new)
```

```python
print(X_new)
print(y_pred)
```

```python
plt.scatter(X,y, label='original data point')
plt.plot(X_new, y_pred, "r-", label='LinearRegession')
plt.xlabel('X variable')
plt.ylabel('Y variable')
plt.title('Linear Regression Plot')
plt.legend()
plt.show()
```

```python
mi=model.intercept_
mc=model.coef_
```
