```python
import numpy as np

# Define products as a dictionary with unique IDs
products = {
    1: {"name": "Tea", "price": 2200, "stock": 5},
    2: {"name": "Milk", "price": 1800, "stock": 8},
    # Add other products here...
}

# Define users as a dictionary with usernames as keys
users = {
    "Ahmed": {"amount": 5000},
    "Blessing": {"amount": 10000},
    # Add other users here...
}

def display_products():
    print("Available Products:")
    for product_id, product in products.items():
        print(f"{product_id}. {product['name']} - N{product['price']} (Stock: {product['stock']})")

def buy_products(user):
    try:
        num_products = int(input("How many products would you like to add to your cart? "))
        total_cost = 0

        for _ in range(num_products):
            product_id = int(input("Enter the product ID: "))
            quantity = int(input("Enter the quantity: "))

            if product_id in products and products[product_id]["stock"] >= quantity:
                total_cost += products[product_id]["price"] * quantity
                products[product_id]["stock"] -= quantity
            else:
                print("Invalid product ID or insufficient stock.")

        if total_cost > user["amount"]:
            print("Insufficient funds.")
        else:
            user["amount"] -= total_cost
            print(f"Purchase successful! Remaining balance: N{user['amount']}")
    except ValueError:
        print("Invalid input. Please enter valid numbers.")

if __name__ == "__main__":
    username = input("Please login with your username: ")
    user = users.get(username)

    if user:
        print(f"Welcome, {username}!")
        print(f"Your account balance is N{user['amount']}.")

        while True:
            print("\nMenu:")
            print("1. See Available Products")
            print("2. Buy Products")
            print("3. Quit")

            choice = input("Enter your choice: ")
            if choice == "1":
                display_products()
            elif choice == "2":
                buy_products(user)
            elif choice == "3":
                print("Thank you for using our service!")
                break
            else:
                print("Invalid choice. Please select a valid option.")
    else:
        print("User not found. Please check your username.")
```

```python
import time

def display_products(product_list):
    print("Available Products:")
    for idx, product in enumerate(product_list, start=1):
        print(f"{idx}. {product['name']} - N{product['price']} (Stock: {product['stock']})")

def add_to_cart(products, shopping_cart, total_cost):
    num_products_to_add = int(input("How many products would you like to add to your cart? "))
    for _ in range(num_products_to_add):
        product_choice = input("Enter the product number: ")
        if product_choice.isdigit():
            product_choice = int(product_choice)
            if 1 <= product_choice <= len(products):
                selected_product = products[product_choice - 1]
                if selected_product["stock"] > 0:
                    shopping_cart.append(selected_product)
                    total_cost += selected_product["price"]
                    selected_product["stock"] -= 1
                else:
                    print("Sorry, this product is out of stock.")
            else:
                print("Invalid product number.")
        else:
            print(f'Please enter the corresponding number for {product_choice}')
    return total_cost

def checkout_process(total_cost, user):
    print("\n...Calculating total cost...\n")
    time.sleep(2)
    print(f"Total Cost: N{total_cost}")
    checkout_option = input("Would you like to proceed?\n1. Yes\n2. No\n")
    if checkout_option == "1":
        if total_cost > user['amount']:
            outstanding_balance = total_cost - user['amount']
            print(f"Insufficient account balance.\nOutstanding N{outstanding_balance} can be added to your tab")
            proceed_response = input('\nWould you like to proceed?\n1. Yes\n2. No\nYour Response: ').lower()
            if proceed_response == '1' or proceed_response == 'yes':
                print(f'A negative balance of N{outstanding_balance} has been added to your customer card\nThanks for shopping with us.')
            elif proceed_response == '2' or proceed_response == 'no':
                print('Purchase failed due to insufficient balance')
            else:
                print('Invalid response. Action denied.')
        else:
            remaining_balance = user['amount'] - total_cost
            print(f"Bye for now, {user['name']}\nRemaining Account Balance: N{remaining_balance}\nThank you for shopping with us.")
    elif checkout_option == "2":
        print('\nQuitting app...\n')
        time.sleep(2)
        print(f"Bye for now, {user['name']}. Come back soon.\n")
    else:
        print("Invalid input. Please enter a valid option.")

def buy_products(products, user):
    # Shopping cart and total cost initialization
    shopping_cart = []
    total_cost = add_to_cart(products, shopping_cart, 0)
    checkout_process(total_cost, user)

def main():
    # List of products with their details
    products = [
        {"name": "Tea", "price": 2200, "stock": 5},
        {"name": "Milk", "price": 1800, "stock": 8},
        {"name": "Sugar", "price": 550, "stock": 5},
        {"name": "Bread", "price": 750, "stock": 8},
        {"name": "Detergent", "price": 850, "stock": 5},
        {"name": "Bathing Soap", "price": 750, "stock": 8},
        {"name": "Coca Cola", "price": 250, "stock": 5},
        {"name": "Body Spray", "price": 2250, "stock": 8},
        {"name": "Tooth Brush", "price": 250, "stock": 5},
        {"name": "Tooth Paste", "price": 750, "stock": 8},
    ]

    # List of users with their account balances
    user_accounts = [
        {"name": "Ahmed", "amount": 5000},
        {"name": "Blessing", "amount": 10000},
        {"name": "Pelumi", "amount": 15000},
        {"name": "Yemi", "amount": 20000},
    ]

    # Ask the user to login with their username
    username_input = input("Please login with your username: ")
    user_matches = [user for user in user_accounts if user["name"] == username_input]

    # Check if the user is in the database and print their name and balance
    if user_matches:
        current_user = user_matches[0]
        print(f"Welcome {current_user['name']}!")
        print(f"Your account balance is {current_user['amount']}.")

        # Ask the user what action they would like to perform
        user_selection = input(
            "What tasks would you like to perform?\n1. See Available Products.\n2. Buy Products.\n3. Quit\n"
        )

        if user_selection == "1":
            display_products(products)
            proceed_option = input(
                "What would you like to do next?: \n1. Buy Products.\n2. Quit.\n"
            )
            if proceed_option == "1":
                buy_products(products, current_user)
            elif proceed_option == "2":
                print('\nQuitting app...\n')
                time.sleep(2)
                print(f"Bye for now, {current_user['name']}. Come back soon.\n")
            else:
                print(f"Invalid input {current_user['name']}\nPlease start over.")
        elif user_selection == "2":
            buy_products(products, current_user)
        elif user_selection == "3":
            print('\nQuitting app...\n')
            time.sleep(2)
            print(f"Bye for now, {current_user['name']}. Come back soon.\n")
        else:
            print("Invalid input. Please enter a valid option.")
    else:
        print("Wrong input, please enter the correct username")

if __name__ == "__main__":
    main()
```
