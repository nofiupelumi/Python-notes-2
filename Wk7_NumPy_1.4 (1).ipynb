## Simple Analysis to Perform on the Dataset

### Get Avg IMDB_Rating

```python
# get all IMDB_Ratings here
imdb_ratings = []

# read file
with open(new_file, 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        # append values. of imdb_rating to the list
        imdb_ratings.append(float(row['IMDB_Rating']))
        
    
print(imdb_ratings)
```

```python
# convert the list to a np array
imdb_ratings_array = np.array(imdb_ratings)
imdb_ratings_array
```

```python
type(imdb_ratings)
```

```python
imdb_ratings_array.dtype
```

```python
type(imdb_ratings_array)
```

```python
# calc the mean of all the movie ratings
avg_imdb_ratings = np.mean(imdb_ratings_array)
print(f"The mean of all {len(imdb_ratings_array)} movies' imdb_ratings is {round(avg_imdb_ratings, 2)}")
```

### Get the Genre Distribution

```python
# create a list to save all movie genres
movie_genres = []

# read file
with open(new_file, 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        # append values. of genre to the genre list
        movie_genres.append(row['Genre'])
        
movie_genres
```

```python
# the np.unique() mtd; this flattens an array and returns all unique values in it.

# count the occurence of each genre
unique_genres, genre_counts = np.unique(movie_genres, return_counts=True)

unique_genres
```

```python
len(unique_genres) # so we see that there are 202 unique set of movie genres
```

```python
# create a dict to map the unique genres and their total count
genre_distribution = dict(zip(unique_genres, genre_counts))
print(f"Movie Genre Distribution: \n {genre_distribution}")
```
